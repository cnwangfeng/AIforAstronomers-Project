## 载入示例数据

这里使用一个简化的太阳系行星数据表。它足够小，适合在入门阶段练习文件读取和基本统计。

```python
import csv
from pathlib import Path

data_path = Path("../../data/small/planetary_orbits_demo.csv")
with data_path.open(newline="", encoding="utf-8") as f:
    rows = list(csv.DictReader(f))

planet_names = [row["planet"] for row in rows]
orbital_periods = [float(row["orbital_period_years"]) for row in rows]
semi_major_axes = [float(row["semi_major_axis_au"]) for row in rows]

print("行星数量:", len(rows))
print("前三个行星:", planet_names[:3])
```

## 变量与数据类型

在科学编程里，我们经常把一个数据列读成 Python 列表，再逐步转换成更适合后续分析的数据结构。

```python
example_planet = planet_names[2]
example_period = orbital_periods[2]

print(example_planet, type(example_planet))
print(example_period, type(example_period))
```

## 列表、索引与切片

列表适合保存一串有顺序的数据。索引从 0 开始，切片可以帮助我们快速查看局部内容。

```python
print("第一个轨道周期:", orbital_periods[0])
print("前四个轨道周期:", orbital_periods[:4])
print("最后两个半长轴:", semi_major_axes[-2:])
```

## 字典的一个典型用途

很多科学任务都需要把“名称”映射到“数值”。字典正适合这种键值关系。

```python
period_by_planet = {
    row["planet"]: float(row["orbital_period_years"])
    for row in rows
}

print("Earth 的轨道周期:", period_by_planet["Earth"])
print("Jupiter 的轨道周期:", period_by_planet["Jupiter"])
```

## 一个最小科学计算示例

开普勒第三定律告诉我们，轨道周期和半长轴之间存在近似关系。这里先不追求精确拟合，只做一个入门级的数量关系比较。

```python
kepler_ratio = []
for axis, period in zip(semi_major_axes, orbital_periods):
    kepler_ratio.append(period**2 / axis**3)

mean_ratio = sum(kepler_ratio) / len(kepler_ratio)
print("平均 P^2 / a^3:", round(mean_ratio, 3))
print("前四个比值:", [round(value, 3) for value in kepler_ratio[:4]])
```
