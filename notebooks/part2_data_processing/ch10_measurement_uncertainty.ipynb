# 第10章 科学数据、误差与单位

本 notebook 用一个最小的视差测量示例说明单位、距离估计和误差传播。

## 学习目标

- 理解测量值、单位和不确定度需要一起记录。
- 学会从视差估计距离。
- 通过一个简化例子理解误差传播的量级。

```python
parallax_mas = 10.0
parallax_uncertainty_mas = 0.2

distance_pc = 1000.0 / parallax_mas
distance_uncertainty_pc = 1000.0 * parallax_uncertainty_mas / (parallax_mas ** 2)

print("distance_pc =", round(distance_pc, 2))
print("distance_uncertainty_pc ~=", round(distance_uncertainty_pc, 2))
```

## 测量模型：真实值、偏差和噪声

正文中把最小测量模型写成 $x_{obs}=x_{true}+b+\epsilon$。下面用一个小数值例子区分系统偏差和随机噪声：偏差会整体推开结果，随机噪声则在重复测量中上下波动。

```python
true_parallax_mas = 10.0
calibration_bias_mas = 0.08
noise_terms_mas = [-0.12, 0.04, 0.16, -0.05]

observed_parallaxes = [
    true_parallax_mas + calibration_bias_mas + noise
    for noise in noise_terms_mas
]

print('observed parallaxes:', [round(value, 3) for value in observed_parallaxes])
print('mean observed parallax:', round(sum(observed_parallaxes) / len(observed_parallaxes), 3))
print('true parallax:', true_parallax_mas)
```

## 重复测量与加权平均

如果每次测量的不确定度不同，就不应该简单把它们等权平均。最小加权平均使用 $w_i=1/\sigma_i^2$，不确定度小的测量权重更大。

```python
measurements = [
    {'parallax_mas': 10.10, 'sigma_mas': 0.30},
    {'parallax_mas': 9.95, 'sigma_mas': 0.20},
    {'parallax_mas': 10.20, 'sigma_mas': 0.50},
]

simple_mean = sum(item['parallax_mas'] for item in measurements) / len(measurements)
weights = [1.0 / (item['sigma_mas'] ** 2) for item in measurements]
weighted_mean = sum(weight * item['parallax_mas'] for weight, item in zip(weights, measurements)) / sum(weights)
weighted_uncertainty = 1.0 / (sum(weights) ** 0.5)

print('simple mean parallax:', round(simple_mean, 3), 'mas')
print('weighted mean parallax:', round(weighted_mean, 3), 'mas')
print('weighted uncertainty:', round(weighted_uncertainty, 3), 'mas')
```

## 为什么单位不能丢

同样的数字 `10.0`，如果单位是毫角秒，它表示一个比较近的恒星；如果误写成角秒，距离就会差上三个数量级。

```python
distance_if_arcsec = 1.0 / 10.0
distance_if_mas = 1000.0 / 10.0

print("If treated as arcsec:", distance_if_arcsec, "pc")
print("If treated as mas:", distance_if_mas, "pc")
```
