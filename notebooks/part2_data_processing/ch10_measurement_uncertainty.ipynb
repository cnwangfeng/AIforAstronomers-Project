# 第10章 科学数据、误差与单位

本 notebook 用一个最小的视差测量示例说明单位、距离估计和误差传播。

## 学习目标

- 理解测量值、单位和不确定度需要一起记录。
- 学会从视差估计距离。
- 通过一个简化例子理解误差传播的量级。

```python
parallax_mas = 10.0
parallax_uncertainty_mas = 0.2

distance_pc = 1000.0 / parallax_mas
distance_uncertainty_pc = 1000.0 * parallax_uncertainty_mas / (parallax_mas ** 2)

print("distance_pc =", round(distance_pc, 2))
print("distance_uncertainty_pc ~=", round(distance_uncertainty_pc, 2))
```

## 为什么单位不能丢

同样的数字 `10.0`，如果单位是毫角秒，它表示一个比较近的恒星；如果误写成角秒，距离就会差上三个数量级。

```python
distance_if_arcsec = 1.0 / 10.0
distance_if_mas = 1000.0 / 10.0

print("If treated as arcsec:", distance_if_arcsec, "pc")
print("If treated as mas:", distance_if_mas, "pc")
```
